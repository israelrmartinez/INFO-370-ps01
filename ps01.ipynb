Hint: use the split() method to get words: "I have no clue".split() -\>

Hint 2: convert everything to lower case 2

Hint 3: use triple quotes """ to create long multi-line strings

Hint 4: proceed as you did with lists: first create an empty dict of
word counts, thereafter loop over words, and increase the count for each
word. But you have to handle the words that are not yet in the dict
somehow.

In \[14\]:

    longString = """Your code should run over individual 
    words and increase the counter for the words"""
    # Hint 1
    words = longString.split()
    # Hint 2
    words = [word.lower() for word in words]

    wordCounts = {}
    for word in words:
        if word in wordCounts.keys():
            wordCounts[word] += 1
        else:
            wordCounts[word] = 1
    wordCounts

    def convertTime(time):
        minute = (time % 100) / 60
        hour = time // 100
        newTime = hour + minute
        return newTime
